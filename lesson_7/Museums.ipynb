```python
sunflower_ids = res['objectIDs']
```

```python
dfs = []
for i in sunflower_ids:
    r_test = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}')
    obj_test = r_test.json()
    if all(field in obj_test for field in ['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate', 'period', 'objectName', 'culture']):
        df = pd.DataFrame.from_dict(obj_test, orient='index') \
        .loc[['objectID', 'title', 'artistDisplayName', 'department', 'objectBeginDate', 'objectEndDate', 'period', 'objectName', 'culture']]
        df = df.transpose()
        df = df.set_index('objectID')
        dfs.append(df)
```

```python
result_df = pd.concat(dfs)
```
