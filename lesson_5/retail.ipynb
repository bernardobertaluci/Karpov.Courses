```python
order_per_users = retail.query("Country == 'Germany'").groupby('CustomerID')['InvoiceNo'].nunique()
```

```python
N = np.percentile(order_per_users, 80)
```

```python
germany_top = order_per_users[order_per_users > N].index
```

```python
germany_top
```

```python
top_retail_germany = retail[retail['CustomerID'].isin(germany_top)]
```

```python
top_retail_germany.groupby('StockCode', as_index=False)['InvoiceNo'].nunique().sort_values('InvoiceNo',ascending=False)
```

```python
retail = retail.assign(Revenue = retail.UnitPrice * retail.Quantity)
```

```python
retail
```

```python
top_5 = retail.groupby('InvoiceNo', as_index=False)['Revenue'].sum().sort_values('Revenue',ascending=False).head(5)
```
