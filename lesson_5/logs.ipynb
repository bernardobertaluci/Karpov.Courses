```python
import pandas as pd
import os
import re
```

```python
sales_data = []
sales_files = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-d-alekseev-38/shared/homeworks/python_ds_miniprojects/5_subsid/subsid/'
```

```python
for file in os.listdir(sales_files):
    pattern = re.compile('log')
    if file.startswith('tm'):
        df = pd.read_csv(os.path.join(sales_files, file), sep=';')
        df = df[['FILIAL_ID', 'SUBS_ID', 'PROD_ID', 'ACT_DTTM']]
        sales_data.append(df.dropna())
    elif pattern.findall(file):
        logs = pd.read_csv(os.path.join(sales_files, file), sep=';')
```

```python
sales = pd.concat(sales_data)
```

```python
sales
```

```python
sales.loc[~sales['SUBS_ID'].str.startswith('id'), 'SUBS_ID'] = 'id' + sales.loc[~sales['SUBS_ID'].str.startswith('id'), 'SUBS_ID']
```

```python
sales
```

```python
logs
```

```python
logs['DURATION'] = pd.to_datetime(logs['END_DTTM'], format='%d-%m-%Y %H:%M') - pd.to_datetime(logs['START_DTTM'], format='%d-%m-%Y %H:%M')
logs
```

```python
logs = logs[logs['DURATION'] >= pd.Timedelta(minutes=5)]
```

```python
logs
```

```python
data = pd.merge(sales, logs, on=['PROD_ID', 'SUBS_ID'], how='inner')
```
