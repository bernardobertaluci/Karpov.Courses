```python
import pandas as pd
```

```python
bookings = pd.read_csv('~/shared/homeworks/python_ds_miniprojects/2/bookings.csv', sep=';')
```

```python
bookings_head = bookings.head(7)
```

```python
renaming_dict = {}
for col_name in bookings.columns:
    new_col_name = col_name.lower().replace(' ', '_')
    renaming_dict[col_name] = new_col_name
bookings.rename(columns=renaming_dict, inplace=True)
```

```python
bookings.head()
```

```python
bookings.groupby('hotel', as_index=False).agg({'stays_total_nights' : 'mean'}).round(2)
```

```python
count = len(bookings[bookings['assigned_room_type'] != bookings['reserved_room_type']])
print(count)
```

```python
bookings.head(20)
```

```python
bookings.agg({'adults' : 'mean', 'children' : 'mean', 'babies' : 'mean'})
```

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
bookings.groupby('hotel', as_index=False).agg({'total_kids' : 'mean'}).round(2)
```

```python
bookings['has_kids'] = bookings['total_kids'] > 0
```

```python
bookings.head(5)
```

```python
all_without_kids = bookings.query('has_kids == False').shape[0]
```

```python
no_canceled_without_kids = bookings.query('has_kids == False').agg({'is_canceled' : 'sum'})
```

```python
outflow_without_kids = (all_without_kids - no_canceled_without_kids) / all_without_kids
print(100 - (outflow_without_kids * 100).round(2))
```

```python
all_with_kids = bookings.query('has_kids == True').shape[0]
```

```python
no_canceled_with_kids = bookings.query('has_kids == True').agg({'is_canceled' : 'sum'})
```

```python
outflow_with_kids = (all_with_kids - no_canceled_with_kids) / all_with_kids
print(100 - (outflow_with_kids * 100).round(2))
```
