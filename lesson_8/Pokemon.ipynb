```python
import pandas as pd
```

```python
df = pd.read_csv('Pokemon.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace(' ', '_').str.replace('.', '').str.lower()
df = df.rename(columns={"#": "id"})
```

```python
df.head()
```

```python
legends = df.groupby(['generation', 'legendary'])['legendary'].count().to_frame()
```

```python
legends = legends.rename(columns={"legendary": "legendary_count"})
```

```python
legends
```

```python
legends_unstacked = legends.unstack()
```

```python
legends_unstacked
```
