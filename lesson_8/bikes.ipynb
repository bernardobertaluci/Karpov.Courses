```python
df1 = pd.read_csv('bikes_q2_sample_apr.csv')
df2 = pd.read_csv('bikes_q2_sample_may.csv')
df3 = pd.read_csv('bikes_q2_sample_jun.csv')
df4 = pd.read_csv('bikes_q3_sample_july.csv')
df5 = pd.read_csv('bikes_q3_sample_aug.csv')
df6 = pd.read_csv('bikes_q3_sample_sep.csv')
df7 = pd.read_csv('bikes_q4_sample_oct.csv')
df8 = pd.read_csv('bikes_q4_sample_nov.csv')
df9 = pd.read_csv('bikes_q4_sample_dec.csv')
```

```python
df_concat = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9])
```

```python
df_concat.head()
```

```python
df_concat.start_time = pd.to_datetime(df_concat.start_time)
```

```python
df_concat.set_index('start_time', inplace=True)
```

```python
df_concat.index
```

```python
df_concat_group = df_concat.groupby('usertype').resample('D').count()
df_concat_group
```

```python
filtered_df = df_concat_group.loc[('Customer', ), 'trip_id'] > df_concat_group.loc[('Subscriber', ), 'trip_id']
```

```python
filtered_df = filtered_df.loc[filtered_df]
```

```python
bikes_summer = df_concat.loc['2018-06-01': '2018-08-31']
bikes_summer
```

```python
top_destination = bikes_summer['to_station_name'].value_counts().index[0]
```

```python
top_destination
```

```python
bikes_summer_top = bikes_summer[bikes_summer['to_station_name'] == top_destination]
```

```python
daily_trips = bikes_summer_top.resample('D').count()
```

```python
bad_day = daily_trips['trip_id'].idxmin().strftime('%Y-%m-%d')
```

```python
bad_day
```

```python
weekend_trips = bikes_summer.loc[bikes_summer.index.weekday.isin([6])]
weekday_trips = bikes_summer.loc[~bikes_summer.index.weekday.isin([5, 6])]
```

```python
weekend_destination_counts = weekend_trips.groupby('to_station_name').count()['trip_id']
weekday_destination_counts = weekday_trips.groupby('to_station_name').count()['trip_id'].sort_values(ascending=False)
```
